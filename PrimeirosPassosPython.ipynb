# Operadores no Python

O Python nos fornece uma variedade de operadores para realizarmos diferentes operações com dados e variáveis. Os operadores desempenham um papel fundamental para a programação com Python, pois são utilizados na construção de expressões lógicas, aritméticas, associativas, atribuir valores a variaveis e compara-los.

No Python temos diferentes tipos de operadores:
* Operadores aritméticos
* Operadores de atribuição
* Operadores de comparação
* Operadores lógicos
* Operadores de identidade
* Operadores de associação

#### Operadores aritméticos

Com os operadores artiméticos podemos realizar operações matemáticas, como adição, subtração, multiplicação, divisão entre outras. Eles nos permitem fazer cálculos numéricos em Python. Aqui estão os principais operadores aritméticos em Python:

Screenshot_2.png

````python
numero1 = 10
numero2 = 15

# Utilizando Operadores de adição
soma = numero1 + numero2
# Utilizando Operadores de subtração
sub = numero1 - numero2
# Utilizando Operadores de multiplicação
mult = numero1 * numero2
# Utilizando Operadores de divisão
div = numero1 / numero2
# Utilizando Operadores de dupla divisão
div2 = numero1 // numero2
# Utilizando Operadores de módulo 
mod = numero1 % numero2
# Utilizando Operadores Exponenciação
exp = numero1 ** numero2

print(f"soma = {soma}")
print(f"subtração = {sub}")
print(f"multiplicação = {mult}")
print(f"divisão = {div}")
print(f"dupla divisão = {div2}")
print(f"módulo = {mod}")
print(f"Exponenciação = {exp}")
````

#### Operadores de atribuição

Os operadores de atribuição atribuem valor a uma variavel. Se você tem alguma experiência em programação, certamente já usou operadores de atribuição antes. O operador de atribuição é igual a "=", mas também existem operadores compostos que combinam a atribuição com uma operação aritmética.

Screenshot_3.png

````python
#utilizando alguns operadores de atribuição
num1 = 5
print(f"A primeira atribuição num1 = {num1}")
num1 += 5
print(f"A segunda atribuição num1 = {num1}")
num1 *= 2
print(f"A Terceira atribuição num1 = {num1}")
````

#### Operadores de comparação

Os operadores de comparação são usados para comparar valores e retornar, com base na condição, se o resultado da comparação é verdadeira ou falsa (TRUE ou FALSE). Os principais operadores de comparação são:

- Igual (==)
- Diferente (!=)
- Maior que (>)
- Menor que (<)
- Maior ou igual (>=)
- Menor ou igual (<=)

````python
num1 = 5
num2 = 6
num3 = 3

# Verificando se os valores são iguais
print(f"A variavel num1 é igual a variavel num2? {num1 == num2}")
# Verificando se os valores são diferentes
print(f"A variavel num1 é diferente da variavel num2? {num1 != num2}")
# Verificando se num1 é maior que num2
print(f"A variavel num1 é maior que a variavel num2? {num1 > num2}")
# Verificando se num1 é menor que num2
print(f"A variavel num1 é menor que a variavel num2? {num1 < num2}")
# Verificando se num3 é maior ou igual a num2
print(f"A variavel num3 é maior ou igual a variavel num2? {num3 >= num2}")
# Verificando se num 3 é menor ou igual a num 2
print(f"A variavel num3 é menor ou igual a variavel num2? {num3 <= num2}")
````

#### Operadores de Identidade

Os operadores de identidade servem para comparar se um objeto ocupa a mesma posição na memória, o que significa que se trata do mesmo objeto. Para realizar essas comparações, utilizamos os seguintes operadores de identidade:

# Controle de fluxo

Em Python, o controle de fluxo é administrado por meio de estruturas condicionais e loops. As estruturas condicionais possibilitam a execução do código com base em determinadas condições, enquanto os loops permitem a repetição contínua de trechos de código.

#### Estruturas condicionais

As estruturas condicionais permitem que um programa execute diferentes blocos de código com base nas condições especificadas, tornando a execução do programa mais dinâmica e adaptativa às diferentes situações e entradas. Portanto, quando estamos falando de estruturas condicionais, podemos pensar nas condições que ocorrem derivadas de um certo fato, por exemplo:

1. Se você tiver menos de 12 anos, você é uma criança.
2. Se você tiver entre 12 e 17 anos, você é um adolescente.
3. Se você tiver entre 18 e 64 anos, você é um adulto.

Para colocarmos essas condições em um codigo de python, vamos usar as seguintes palavra-chaves:
* **if:**  É usada para testar uma condição. Se a condição for verdadeira, o bloco de código será executado.
* **elif:** Testa uma condição adicional se a condição anterior (if ou outro elif) for falsa.
* **else:** Define um bloco de código que será executado se todas as condições anteriores (if e elif) forem falsas.

````python
# Criando uma estrutura de decisão baseada no exemplo dado acima

idade = 25

if idade < 12:
    print("Você é uma criança")
elif (idade >= 12) and (idade < 18): # Utilizando os operadores lógicos
    print("Você é um adolescente")
elif 18 <= idade <= 64: #outra alternativa de como escrever
    print("você é um adulto")
else:
    print("Você é um idoso")
````
