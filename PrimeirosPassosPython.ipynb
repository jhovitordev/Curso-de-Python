#### **Comentários**

Uma boa prática na programação é a utilização de comentários para explicar partes do código, pois, conforme formos avançando no curso de Python e na programação, iremos perceber a necessidade de dividir o nosso código em seções. Para deixarmos nosso código mais legível, podemos utilizar dos comentários como uma alternativa para explicar cada uma dessas seções.

*Os comentários são ignorados durante a execução do código.*

Podemos utilizar duas formas diferentes para comentar em um código em Python:

* Comentários de linha única: Utilizamos os símbolo "#" para criarmos um comentário que ocupa apenas uma linha.
* Comentários de múltiplas linhas: Utilizamos 3 aspas simples (''') ou duplas (""") para criar comentários que ocupam varias linhas.

````python
# aqui um exemplo de comentários single line com a "#"

'''
Autor: João Vitor
Esse exemplo de multi line
'''

print("Os comentários são ignorados pelo interpretador")
````

#### **Tipo de dados**

Quando estamos lidando com Python, precisamos compreender comos esses dados são trabalhados por ele. Por exemplo, quando falamos que iremos fazer um cálculo, conseguimos entender rapidamente que vamos trabalhar com numeros. Já para o python, precisamos especificar claramente para ele o que iremos usar.

1.   Texto : O Python entende como String;

#### **Variaveis**


As variáveis são criadas para armazenar informações e serem manipuladas em um programa. Então, quando vamos pensar em variaveis, podemos entende-las como se fossem caixinhas nomeadas que você guarda informações, e sempre que você abre essa caxinha, você consegue ver o conteudo dela.

Para criarmos uma variavel e atribuir um valor para ela em python é muito fácil, basta escrever o nome dela e atribuir o valor que deseja.

*Devemos nos atentar sempre para não usarmos palavras reservadas do python como nome de variáveis*

Variables.png

Uma boa pratica é sempre rotular dados com variaveis que tenham nomes bem descritivos, para que nossos programas possam ser entendidos mais claramente pelo leitor e por nós mesmos.


**Modificando Tipo de Dados**

Podemos pegar um tipo de dado e transforma-lo em outro com as seguintes funções:

*   int()
*   str()
*   float()



````python
#criando variaveis
nome = 'João Vitor'
idade = 25
altura = 1.75

#usando as variaveis
print(nome)
print(idade)
print(altura)

#modificando tipo dos dados
valorModificado = str(idade)
print(f"O tipo da variavel é: {type(valorModificado)}")
print(f"O valor da variavel é: " + valorModificado) # python só consegue concatenar entre strings
````

#### **Input**

Utilizamos a função input() quando precisamos que o usuário passe ao programa algum tipo de dado. Podemos utilizala da seguinte forma:



```
nome_usuario = input("Qual seu nome?")
```

Aqui nós realizamos os seguintes passos:


*   Declaramos a variavel "nome_usuario"
*   Chamamos a função "input()" que solicita o nome da pessoa printando a mensagem na tela

Quando o usuario fornecer o valor solicitado, estre valor será guardado na variavel "nome_usuario" como uma string.

*Até aqui, você já conseguiu capturar dados, armazenar esses dados e apresentar eles na tela*


````python
nome_usuario = input("Qual seu nome?")
idade_usuario = input("Qual sua idade?")
print( nome_usuario + ' tem ' + idade_usuario + " anos de idade")
````

# Operadores no Python

O Python nos fornece uma variedade de operadores para realizarmos diferentes operações com dados e variáveis. Os operadores desempenham um papel fundamental para a programação com Python, pois são utilizados na construção de expressões lógicas, aritméticas, associativas, atribuir valores a variaveis e compara-los.

No Python temos diferentes tipos de operadores:
* Operadores aritméticos
* Operadores de atribuição
* Operadores de comparação
* Operadores lógicos
* Operadores de identidade
* Operadores de associação

#### Operadores aritméticos

Com os operadores artiméticos podemos realizar operações matemáticas, como adição, subtração, multiplicação, divisão entre outras. Eles nos permitem fazer cálculos numéricos em Python. Aqui estão os principais operadores aritméticos em Python:

Screenshot_2.png

````python
numero1 = 10
numero2 = 15

# Utilizando Operadores de adição
soma = numero1 + numero2
# Utilizando Operadores de subtração
sub = numero1 - numero2
# Utilizando Operadores de multiplicação
mult = numero1 * numero2
# Utilizando Operadores de divisão
div = numero1 / numero2
# Utilizando Operadores de dupla divisão
div2 = numero1 // numero2
# Utilizando Operadores de módulo 
mod = numero1 % numero2
# Utilizando Operadores Exponenciação
exp = numero1 ** numero2

print(f"soma = {soma}")
print(f"subtração = {sub}")
print(f"multiplicação = {mult}")
print(f"divisão = {div}")
print(f"dupla divisão = {div2}")
print(f"módulo = {mod}")
print(f"Exponenciação = {exp}")
````

#### Operadores de atribuição

Os operadores de atribuição atribuem valor a uma variavel. Se você tem alguma experiência em programação, certamente já usou operadores de atribuição antes. O operador de atribuição é igual a "=", mas também existem operadores compostos que combinam a atribuição com uma operação aritmética.

Screenshot_3.png

````python
#utilizando alguns operadores de atribuição
num1 = 5
print(f"A primeira atribuição num1 = {num1}")
num1 += 5
print(f"A segunda atribuição num1 = {num1}")
num1 *= 2
print(f"A Terceira atribuição num1 = {num1}")
````

#### Operadores de comparação

Os operadores de comparação são usados para comparar valores e retornar, com base na condição, se o resultado da comparação é verdadeira ou falsa (TRUE ou FALSE).
